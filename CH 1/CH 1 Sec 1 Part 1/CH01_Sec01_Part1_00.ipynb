Pada bab ini kita akan mempelajari metode Gaussian Elimination dan Back subtitution untuk menyelesaikan persamaan sistem linear. Mari mulai dengan sebuah contoh

Contoh #1 :
Carilah nilai untuk sistem x, y, dan z dari :

    x  + y  + z = 6
    2x - y  + z = 3
    3x + 2y - z = 4

kita akan mulai mencari nilai x, y, dan z dengan metode eliminasi gaussian yang penyelesaian analitiknya dapat dilihat di : url

adapun mencari nilai x,y, dan z:

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

plt.rcParams['figure.figsize'] = [14,8]
plt.rcParams.update({'font.size':18})

# Cara Pertama
A = np.array([[1., 1, 1, 6], 
              [2, -1, 1, 3],
              [3, 2, -1, 4]])
r1 = A[0,:]
r2 = A[1,:]
r3 = A[2,:];
A = np.array([r1,
              r2-2*r1,
              r3-3*r1])
r1 = A[0,:]
r2 = A[1,:]
r3 = A[2,:]
A = np.array([r1,
              r2,
              r3-(1/3)*r2])
print(A)
```

Dari hasil program ini kita dapatkan bentuk matrix A yang baru, dan dapat langsung melakukan Back Substitution 

###### Adapun metode kedua yang jauh lebih cepat untuk mencari nilai x, y, dan z adalah sebagai berikut

```python
## Metode kedua
A = np.array([[1, 1, 1], 
              [2, -1, 1],
              [3, 2, -1]])
B = np.array([6,3,4])
x = np.linalg.solve(A,B)
print(x)
```

Dari hasil ini kita dapatkan nilai sistem untuk persamaan sistem linear ini adalah :
    
    x = 1
    y = 2
    z = 3
    
#### Visualisasi Hasil yang kita dapatkan

```python
x = np.linspace(0,2,20)
y = np.linspace(1,3,20)
X,Y = np.meshgrid(x,y)
Z = 6-X-Y
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(X,Y,Z, facecolor='blue', edgecolor='none',alpha=0.5)
Z = 3-2*X+Y;
ax.plot_surface(X,Y,Z,facecolor='red', edgecolor='none',alpha=0.7)
Z = 3*X+2*Y-4;
ax.plot_surface(X,Y,Z,facecolor='green',alpha=0.7)
ax.plot(1,2,3,'o', markersize=12, markerfacecolor='black')

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

ini merupakan bukti visualisasi yang sangat jelas bahwa titik (1,2,3) berada pada seluruh 3 bidang, sehingga dapat disimpulkan bahwa jawaban kita benar.
